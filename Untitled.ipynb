```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/Users/thomasmac/Downloads/forecast_history.csv")
```

```python
def clean_forecast(column):
    column = column.str.replace('%', '').replace('$', '').replace('I5', '15', regex=False)
    column = pd.to_numeric(column, errors='coerce') / 100  # Convert to float and divide by 100
    return column

# Clean the cols
df['Westpac: 4 year forecast'] = clean_forecast(df['Westpac: 4 year forecast'])
df['Joe Bloggs: 2 year forecast'] = clean_forecast(df['Joe Bloggs: 2 year forecast'])
df['Harry Spent: 5 year forecast'] = clean_forecast(df['Harry Spent: 5 year forecast'])

# Missing Values
df['Westpac: 4 year forecast'].fillna(df['Westpac: 4 year forecast'].mean(), inplace=True)
df['Joe Bloggs: 2 year forecast'].fillna(df['Joe Bloggs: 2 year forecast'].mean(), inplace=True)
df['Harry Spent: 5 year forecast'].fillna(df['Harry Spent: 5 year forecast'].mean(), inplace=True)

# Update
df['Median house price'] = pd.to_numeric(df['Median house price'], errors='coerce')

df
```

```python
df.columns = ['Year', 'Median house price', 'Westpac: 4 year forecast', 'Joe Bloggs: 2 year forecast', 'Harry Spent: 5 year forecast']
```
